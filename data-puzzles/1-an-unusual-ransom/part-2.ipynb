## An unusual ransom (part 2)

#### Can you find the silhouette hidden in this dataset?

- Assume that each feature is normally distributed $\to$ histogram plot for each feature.
- Observe that `feature_2` and `feature_5` are actually not normally distributed.
- Scatter plot these 2.

**Solution: `unicorn`**

```python
from pathlib import Path

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
sns.set()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/hyperc54/data-puzzles-assets/master/visualisation/chal1/chal_visu_1_q2_data.csv')
print(df.shape)
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(18,12))

for i in range(7):
    k = f'feature_{i}'
    plt.subplot(3, 4, i+1)
    plt.hist(df[k], bins=32)
    plt.title(k)
```

```python
plt.scatter(df['feature_2'], df['feature_5'])
```
