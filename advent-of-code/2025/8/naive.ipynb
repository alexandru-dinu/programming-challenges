```python
from collections import defaultdict
from itertools import combinations
from math import prod

from tqdm import tqdm, trange


def dist(a, b):
    return sum((i - j) ** 2 for i, j in zip(a, b))
```

```python
# FILE, ITERS = "example", 10
FILE, ITERS = "input", 1000

with open(FILE) as fp:
    ps = [tuple(map(int, x.split(","))) for x in fp.read().strip().split("\n")]
```

```python
comb = set(combinations(ps, 2))
len(comb)
```

```python
dists = {}
for c in tqdm(comb):
    dists[c] = dist(*c)
```

## Part 1

```python
gs = [{p} for p in ps]
cs = set()

# TODO: Union-Find (DSU)

for _ in trange(ITERS):
    (u, v) = min(dists.keys() - cs, key=dists.get)
    cs |= {(u, v)}

    gu = None
    gv = None
    for g in gs:
        if u in g:
            gu = g
        if v in g:
            gv = g

    if gu and gv:
        if gu == gv:
            continue
        gu |= gv
        del gv
    elif gu and not gv:
        gu.add(v)
    elif not gu and gv:
        gv.add(u)
    else:
        assert False, "at least one group should exist"
```

## Part 2

```python
gs = [{p} for p in ps]
cs = set()

# TODO: Union-Find (DSU)

pbar = tqdm(desc="Merging")

with pbar:
    while True:
        (u, v) = min(dists.keys() - cs, key=dists.get)
        cs |= {(u, v)}

        gu = None
        gv = None
        for g in gs:
            if u in g:
                gu = g
            if v in g:
                gv = g

        if gu and gv:
            if gu == gv:
                continue
            gu |= gv
            del gv
        elif gu and not gv:
            gu.add(v)
        elif not gu and gv:
            gv.add(u)
        else:
            assert False, "at least one group should exist"

        if len(gs) == 1:
            print(">>>", u, v, u[0] * v[0])
            break

        pbar.update(1)
        pbar.set_description(f"{len(gs)=:,d}")
```
